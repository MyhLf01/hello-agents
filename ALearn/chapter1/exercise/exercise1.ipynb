# 分析case主体是否是智能体，可以从多维度解析是哪个类型的智能体
- case1: 一台符合冯·诺依曼结构的超级计算机，拥有高达每秒 2EFlop 的峰值算力

    严格来说不算智能体，因为它没有自主的传感器和执行器，只是执行人类预设的程序。如果非要归类：内部架构角度看是反射智能体，基于预设规则计算；时间角度看是反应式智能体，响应极快；知识角度看是符号主义。

- case2: 特斯拉自动驾驶系统在高速公路上行驶时，突然检测到前方有障碍物，需要在毫秒级做出刹车或变道决策

    内部架构角度是基于模型的智能体——它内部维护着对道路、车辆、障碍物的实时模型，不是简单条件反射。时间角度是反应式（毫秒级响应优先于规划）。知识角度是符号主义（明确的规则和模型）。

- case3: AlphaGo在与人类棋手对弈时，需要评估当前局面并规划未来数十步的最优策略

    内部架构是基于效用和目标的混合智能体——它用蒙特卡洛树搜索（符号推理）来规划数十步，用深度神经网络（亚符号）来评估局面价值，目标是全局最优而非当下最优。时间角度属于规划智能体。知识角度——DeepSeek v4 Pro 也认为是神经符号主义（深度神经网络 + 符号搜索的混合体），纠正我之前的"纯亚符号"判断。

- case4: ChatGPT 扮演的智能客服在处理用户投诉时，需要查询订单信息、分析问题原因、提供解决方案并安抚用户情绪

    内部架构基于目标的智能体——以"解决用户问题并安抚情绪"为目标驱动行为。时间角度是混合式：简单问题即时回复（系统1），复杂问题需要多步推理（系统2）。知识角度是神经符号主义：底层是神经网络，但调用订单系统、应用退款规则时体现符号推理。
