# 智能健身教练设计任务环境
PEAS模型

P（Performance，性能度量）：健身前根据用户目标制定个性计划；健身中根据实时生理数据和动作姿态提供即时指导；健身后给出效果评估和饮食推荐。

E（Environment，环境）：用户的身体数据（心率、血氧、卡路里消耗等）、运动姿态、健身目标、历史训练记录、外部参考资料（运动科学文献、营养学数据库）。

A（Actuators，执行器）：生成训练计划并展示给用户、调用语音接口播报实时指导、根据数据调整计划、生成评估报告和饮食建议。

S（Sensors，传感器）：可穿戴设备数据接口（JSON 解析心率/血氧等）、摄像头捕捉并分析用户运动姿态、用户手动输入的目标和偏好。

环境特性：
- 部分可观察：传感器只能捕捉部分生理指标和姿态，无法感知用户主观感受（如疲劳程度、肌肉酸痛）
- 随机性：用户当天的身体状态、精神状态是随机的，同一训练计划每次执行效果不同
- 动态性：环境随用户运动不断变化（心率上升、动作变形），且用户目标本身也可能随时间调整
- 序列性：当前的训练决策会影响后续的训练效果，不是孤立的一次性决策
