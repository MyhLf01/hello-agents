# 两种方案来处理售后退款申请

## Workflow
设计一套固定流程，按规则层层判断：

优点：确定性高，设计好就不会出错，响应快，成本低（不需要调用 LLM），审计方便（每一步都可追溯）。

缺点：设计复杂，要把大量边界条件提前写好；遇到未覆盖的新情况直接卡住；政策变动需要改动整个工作流，维护成本高；不够灵活，无法处理规则之外的灰色地带。

## Agent
搭建一个智能体系统，让它理解退款政策、分析用户历史行为、评估商品状况，并自主决策是否批准退款：

优点：极其灵活，即使遇到新政策或未预见的情况也能自主推理应对；可以处理模糊的边界 case（如"商品有小瑕疵但影响不大"）；减少人工规则维护成本。

缺点：结果不确定，可能理解错误或判断偏差；LLM API 调用有失败风险；每次决策都需要推理，响应慢且成本高；出了问题很难追溯"为什么拒绝了这个退款"。

## 场景选择

流程比较清晰固定、规则明确的场景（如"7天内、<100元自动退"）适合 Workflow；流程可能变动、或需要处理模糊判断的场景（如"用户说商品和描述不符但没有明确证据"）适合 Agent。

我认为**比较好的方式是 Workflow 中包含 Agent**：用 Workflow 处理 80% 的确定性 case（快速、低成本），把复杂的灰色地带 case 分流给 Agent 做深度判断。这样兼顾了效率与灵活性。
