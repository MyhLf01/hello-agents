# 系统1 vs 系统2 智能体应用场景

参考卡尼曼《思考，快与慢》中的双系统理论：
- **系统1（直觉式）**：快速、自动、不费力、情绪化，基于模式识别。
- **系统2（推理式）**：缓慢、受控、需投入精力、逻辑性，基于分析推理。

对于 AI 智能体来说：
- 系统1 适合**简单反射型**任务：感知→匹配模式→直接行动，毫秒级响应。
- 系统2 适合**深度推理型**任务：感知→分析→规划→多步行动，需要"停下来想一想"。

---

## 哪些应用场景需要"系统1"？

1. **实时异常检测**（如工业质检、网络入侵检测）：摄像头/传感器捕捉数据，瞬间判断是否异常，不能等几秒——生产线等不起。

2. **语音助手唤醒词识别**："Hey Siri" 之类的唤醒词必须在本地、低功耗、极快响应。如果用大模型慢吞吞推理，用户体验极差。

3. **高频交易**：市场数据毫秒级变化，需要预先训练好的策略模型瞬间决断，来不及做深度分析。

4. **推荐系统第一层粗筛**：从几百万商品里快速筛出几百个候选，靠 embedding 相似度匹配（系统1）；精细排序再交给系统2。

## 哪些应用场景需要"系统2"？

1. **医疗诊断辅助**：病人的症状、化验单、影像报告组合在一起，需要逐层推理、鉴别诊断、查阅文献，不能凭"直觉"给结论。

2. **法律合同审查**：逐条分析条款含义、比对案例库、识别潜在风险——每一步都需要仔细推敲。

3. **复杂数学证明**：不是模式匹配，需要探索多条路径、回溯、验证，典型的系统2场景。

4. **第一章的旅行助手智能体**：分析用户请求→调用天气 API→调用景点 API→整合信息→给出建议，多步规划+工具调用，正是系统2。

## 两个系统如何协同工作？

以**智能客服系统**为例：

1. **系统1 做第一层**：用户发消息 → 意图识别（退换货/查询物流/投诉）→ 简单问题（如"我的快递到哪了"）直接查数据库秒回。
2. **系统2 兜底**：复杂问题（如"我收到的商品和描述不符，而且已经过了7天，但我有正当理由"）→ 转给 LLM 智能体进行多步推理，查政策、看历史记录、评估证据、生成回复。

再以**自动驾驶**为例：
- 系统1：车道保持、紧急刹车——端到端神经网络，毫秒级。
- 系统2：导航路径规划——考虑实时路况、施工信息、用户偏好。
- 协同：系统1 保证安全底线，系统2 负责优化决策。系统1 遇到无法处理的模糊场景时，降级报警并移交系统2。

> **核心启示**：真实世界的智能体系统，几乎都是 系统1+系统2 的混合架构。LLM 天然擅长系统2，但成本高、速度慢；小模型/规则擅长系统1。关键的设计问题是：**什么样的任务交给谁，以及两者之间的切换触发条件是什么**。
